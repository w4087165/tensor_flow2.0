```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

# tensor flow2.0 梯度下降 线性回归

```python
# 随机生成1000个点，围绕在y=0.1x+0.3 的直线周围
num_points = 100
vectors_set = []
for i in range(num_points):
    x1 = np.random.normal(0.0,0.55)
    y = x1*0.1 +0.3 + np.random.normal(0.0,0.02)
    vectors_set.append([x1,y])
# 生成一些样本
x_data = [v[0] for v in vectors_set]
y_data = [v[1] for v in vectors_set]
plt.scatter(x_data,y_data,c='orangered')
plt.show()
```

```python

w = tf.Variable(1.)
b = tf.Variable(1.)
# 以估计值y和实际值y_data之间的均方误差
# 采用梯度下降来优化参数
 #迭代200次。
for i in range(1000):
    with tf.GradientTape() as tape:
        tape.watch([w,b])
        logits = w*x_data+b
        loss = tf.reduce_mean(tf.losses.MSE(y_data,logits))

    grads= tape.gradient(loss,[w,b])

    w = w - 0.01*grads[0]
    b = b - 0.01*grads[1]
    
    if i%20 ==0:
        print('sep:{},d_w:{},d_b:{},loss:{}'.format(i,w,b,loss))
```

# 最后迭代后 看到 w趋近与设定的0.1  b趋近于设定的0.3

```python

x_data = np.array(x_data)
y_data = np.array(y_data)
pred_y = w*x_data +b
```

```python
plt.scatter(x_data,y_data)
plt.plot(x_data,pred_y)
plt.show()
```
